```python
#Extractin the records when a team won batting first
batting_first = ipl[ipl['win_by_runs']!=0]
```

```python
#Loking at the head
batting_first.head()
```

```python
#Making a histogram
plt.figure(figsize=(5,7))
plt.hist(batting_first['win_by_runs'])
plt.title("Distribution_of _run")
plt.xlable("runs")
plt.show()
```

```python
#Find out the number of win w.r.t each team after batting first 
batting_first['winner'].value_counts()
```

```python
#Making a bar-plot for top3 team with most wins after batting first 
plt.figure(figsize=(6,4))
plt.bar(list(batting_first['winner'].value_counts()[0:3].keys()),list(batting_first['winner'].value_counts()[0:3]),color=["black","red","orange"])
plt.show()
```

```python
#Making pie chart
plt.figure(figsize=(7,7))
plt.pie(list(batting_first['winner'].value_counts()),labels = list(batting_first['winner'].value_counts().keys()),autopct = '%0.1f%%')
plt.show()
```
